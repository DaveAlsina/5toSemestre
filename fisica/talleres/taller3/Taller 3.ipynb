```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def theta2(t: np.array):
    """t: vector columna"""
    
    return (t**2)/2
```

```python
def radious2(t: np.array):
    
    """t: vector columna"""
    
    return 1 + (np.sin(t*np.pi/180)/2)
                  
```

```python
def circular_movement(theta, radious, t):
    
    angles = theta(t)*np.pi/180
    distance = radious(t)
    
    r_x = distance*np.cos(angles)
    r_y = distance*np.sin(angles)
    
    fig1, ax1 = plt.subplots(1, dpi=400)
    
    ax1.plot(r_x, r_y)
    ax1.set_title("Movimiento circular no uniforme")
    ax1.set_xlabel("Posici贸n en x (mts)")
    ax1.set_ylabel("Posici贸n en y (mts)")
    
    fig2, ax2 = plt.subplots(1)
    ax2.plot(angles, r_x)
    ax2.set_title("Movimiento circular no uniforme")
    ax2.set_xlabel("Angulo (grados)")
    ax2.set_ylabel("Posici贸n en x (mts)")
    
    fig3, ax3 = plt.subplots(1)
    ax3.plot(angles, r_y)
    ax3.set_title("Movimiento circular no uniforme")
    ax3.set_xlabel("Angulo (grados)")
    ax3.set_ylabel("Posici贸n en y (mts)")
    
    
    plt.show()
```

```python
circular_movement(theta2, radious2, np.arange(0, 500, step=0.001))
```
