```python
def generate_dataset(num_nodes: int, nsamples:int, store = True):
    
    for i in range(nsamples):
        
        rand_sparse = create_random_sparse(nodos = num_nodes)
        g = Graph(sparse_mat = rand_sparse)
        nx_g = nx.convert_matrix.from_numpy_array( g.sparse2adj(g.sparse))
        
        print("a:", nx_g)
        
        img  = nx.adjacency_matrix(nx_g, weight = 'weight')
        img  = csr_matrix.todense(img)
        
        plt.imshow(img)
        
        F=nx.floyd_warshall(nx_g)
        #print(F)
        
        T=nx.minimum_spanning_tree(nx_g)
        print(T)
        
```
